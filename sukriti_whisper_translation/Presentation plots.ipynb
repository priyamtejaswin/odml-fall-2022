```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
wer = [21.42297, 15.53726, 10.76960, 8.900819, 50.637057]
latency = [1.02, 1.63, 6.00, 21.07, 32.44]
params = np.array([39, 74, 244, 769, 76])
params = params/2
size = [151097779, 290457431, 967101635, 3055755675, 293000000]
size = np.array(size)/1e6
models = ["whisper tiny", "whisper base", "whisper small", "whisper medium", "emformer"]
print(size)
perf_under_budget = [6000, 3158, 3000, 594, 209]
```

```python
plt.figure(figsize=(6, 4))
#plt.scatter(y = wer, x = latency, s = params, text = models)
for i in range(5):
    x = latency[i]
    y = wer[i]
#     plt.plot(x, y, 'bo')
#     plt.text(x * (1 + 0.01), y * (1 + 0.01) , models[i], fontsize=6)
    plt.scatter(y=y, x=x, s=params[i], label=models[i])

plt.legend()
plt.ylabel("Word Error Rate (in %)")
plt.xlabel("Latency (in seconds)")
plt.title("Word Error Rate vs Latency")
plt.grid()

plt.savefig("WER-Latency.png")
```

```python
plt.figure(figsize=(6, 4))
#plt.scatter(y = wer, x = size, s = params)
for i in range(5):
    x = size[i]
    y = wer[i]
    plt.scatter(y=y, x=x, s=params[i], label=models[i])

plt.legend()
plt.ylabel("Word Error Rate (in %)")
plt.xlabel("Disk Size (in MB)")
plt.title("Word Error Rate vs Disk Size")
plt.grid()
plt.savefig("WER-Size.png")
```

```python
#Latency vs Size
plt.figure(figsize=(6, 4))
#plt.scatter(y = size, x = latency, s = params)
for i in range(5):
    x = latency[i]
    y = size[i]
    plt.scatter(y=y, x=x, s=params[i], label=models[i])

plt.legend()
plt.xlabel("Latency (in seconds)")
plt.ylabel("Disk Size (in MB)")
plt.title("Disk Size vs Latency")
plt.grid()
plt.savefig("Latency-Size.png")
```
